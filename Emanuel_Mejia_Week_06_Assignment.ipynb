```
# Creating a new Dataframe: only GoalKeeper's Data

select_col = ['short_name','nationality','club', 'preferred_foot', 'potential','wage_eur', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_reflexes', 'gk_speed', 'gk_positioning']
col_names = ['Name','Nationality','Club', 'Preferred_foot', 'Potential','Wage_Euros', 'Diving', 'Handling', 'Kicking', 'Reflexes', 'Speed', 'Positioning']

gk_df = df.loc[df['team_position']=="GK"]
gk_df = gk_df[select_col]
gk_df.columns = col_names

gk_df
```

```
chart = alt.Chart(
            gk_df,
            title="Player Salary VS Potential"
        ).mark_bar(
            tooltip = True,
            size = 100,
            fillOpacity= 0.7,
            cornerRadius = 10
        ).encode(
            x=alt.X('Wage_Euros',
                    title = "Wage (€)",
                    bin = alt.Bin(maxbins=5),
                    axis = alt.Axis(titleColor = 'gray',
                                    labelColor = 'gray',
                                    titleFontSize = 15)),
            y=alt.Y('average(Points):Q',
                    title = "Average Skill Points",
                    axis = alt.Axis(titleColor = 'gray',
                                    labelColor = 'gray',
                                    ticks = False,
                                    titleFontSize = 15)),
            color = 'Skill:N',
            order=alt.Order('Skill:N',
                            sort='ascending'
                            ),
            detail = alt.Detail('count()', title = 'Player Count'),
            text = 'Potential'
        ).transform_fold(
            as_=['Skill', 'Points'],
            fold=['Diving', 'Handling', 'Kicking', 'Reflexes', 'Speed', 'Positioning']
        ).configure_view(
            stroke=None
        ).properties(
            title={
              "anchor": "start",
              "text": "Goalkeepers' Skill points by Annual Wage",
              "subtitle": "6 Goalkeeping Skills (measured from 1 to 100)",
              "color": "royalblue",
              "font": "Sans-Serif",
              "fontSize": 30,
              "subtitleColor": "black",
              "subtitleFont" : "Sans-Serif",
              "subtitleFontWeight" : "bold",
              "subtitleFontSize": 17,
            },
            width = 1000,
            height = 600
        ).configure_legend(
            strokeColor='blue',
            fillColor='aliceblue',
            padding=10,
            cornerRadius=10,
        )

chart
```
